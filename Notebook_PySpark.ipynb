```python
## Spark Session
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Day-2-RDDs").getOrCreate()

sc = spark.sparkContext
```

```python
## Create RDD from python List

numbers = [1,2,3,4,5,6]

rdd_numbers = sc.parallelize(numbers)


rdd_numbers.collect()
```

```python
## Map Transformations

rdd_squared = rdd_numbers.map(lambda x:x*x)

rdd_squared.collect()
```

```python
## Filter Transformation

rdd_even = rdd_numbers.filter(lambda x: x%2 ==0)

rdd_even.collect()
```

```python
## Create RDD from a text file

rdd_text = sc.textFile("Files/Sample.txt")

rdd_text.take(5)
```

```python
# Word Count Example

words = rdd_text.flatMap(lambda line:line.split(","))
word_pairs = words.map(lambda w:(w,1))

word_count = word_pairs.reduceByKey(lambda a, b: a+b)

word_count.collect()
```

```python
## RDD -> Dataframe Conversion

data = [
    (1,"A"),
    (2,"B"),
    (3,"C")
]

rdd = sc.parallelize(data)

df= spark.createDataFrame(rdd,["id","value"])

display(df)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Day-8-Joins").getOrCreate()
```

```python
emp_data = [
    (1,"Tanvi",101),
    (2,"Rahul",102),
    (3,"Shiva",103),
    (4,"Amit",None)
]

emp_cols = ["id","name","deptid"]

emp_df = spark.createDataFrame(emp_data,emp_cols)

emp_df.show()
```

```python
dept_data = [
    (101,"HR"),
    (102,"IT"),
    (103,"Finance")
]

dept_cols = ["deptid","Dname"]

dept_df = spark.createDataFrame(dept_data,dept_cols)
dept_df.show()
```

```python
## Inner Join

inner_join = emp_df.join(dept_df,emp_df.deptid == dept_df.deptid,"inner")
inner_join.show()
```

```python
## Left Join

left_join = emp_df.join(dept_df,emp_df.deptid == dept_df.deptid,"left")
left_join.show()
```

```python
## Right Join

right_join = dept_df.join(emp_df,emp_df.deptid == dept_df.deptid,"right")
right_join.show()

```

```python
## full join

full_join = emp_df.join(dept_df,emp_df.deptid == dept_df.deptid,"full")
full_join.show()
```

```python
## join using col name

emp_df.join(dept_df,"deptid","inner").show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Day10-Filtering Data").getOrCreate()
```

```python
data = [
    (1,"Tanvi","IT",6000),
    (2,"Rahul","HR",3000),
    (3,"Kashyap","Finance",2000),
    (4,"Neha","HR",5000),
]

columns = ["id","name","dept","salary"]

df= spark.createDataFrame(data,columns)
```

```python
## Filter() Function

df.filter(df.dept=="IT").show()
```

```python
## Where Function

df.where(df.salary>2000).show()
```

```python
## Multiple Conditions (AND)

df.filter((df.dept=="IT") & (df.salary>2000)).show()
```

```python
# OR
df.filter((df.dept=="IT") | (df.salary>2000)).show()
```

```python
# isin function

df.filter(df.dept.isin("IT","HR")).show()
```

```python
## between

df.filter(df.salary.between(2500,6000)).show()
```

```python
df.filter(df.salary>3000).select("name","salary").show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Day11-Aggregations").getOrCreate()
```

```python
data = [
    (1,"IT",20000),
    (2,"HR",34999),
    (3,"IT",23000),
    (4,"Finance",23030),
    (5,"HR",26000),
    (6,"IT",34000),
]

columns = ["emp_id","dept","sal"]

df = spark.createDataFrame(data,columns)
df.show()
```

```python
## Group BY Basic
df.groupBy("dept").count().show()
```

```python
#Sum Aggregation

df.groupBy("dept").sum("sal").show()
```

```python
#avg 

df.groupBy("dept").avg("sal").show()
```

```python
#Multiple Aggregation

from pyspark.sql.functions import sum, avg

df.groupBy("dept").agg(
    sum("sal").alias("total_sal"),
    avg("sal").alias("avg_sal")
).show()
```

```python
#Count Distinct

from pyspark.sql.functions import countDistinct

df.select(countDistinct("dept")).show()
```

```python
#agg without group by

df.select(sum("sal")).show()
```
